```python
import pandas as pd
import numpy as np
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt
def resample_dataframe(df, original_rate, new_rate):
    # Generate original time indices
    original_time = np.arange(len(df)) / original_rate
    # Generate new time indices
    new_length = int(len(df) * new_rate / original_rate)
    new_time = np.arange(new_length) / new_rate
    # Create a new DataFrame to store the resampled values
    resampled_df = pd.DataFrame(index=new_time)
    for column in df.columns:
        # Create interpolation function for each column
        interp_func = interp1d(original_time, df[column].values, kind='linear', fill_value='extrapolate')
        # Generate new values at the new sampling rate
        resampled_df[column] = interp_func(new_time)
    return resampled_df
# Example DataFrame with some columns (e.g., two 1-second sine waves sampled at 25 Hz)
original_rate = 25  # Hz
new_rate = 30  # Hz
time_index = np.arange(original_rate) / original_rate
data = {
    'Signal1': np.sin(2 * np.pi * 1 * time_index),
    'Signal2': np.cos(2 * np.pi * 1 * time_index)
}
df = pd.DataFrame(data, index=time_index)
# Resample DataFrame
resampled_df = resample_dataframe(df, original_rate, new_rate)
# Plotting the results for visualization
plt.figure(figsize=(10, 5))
for column in df.columns:
    plt.plot(df.index, df[column], 'o-', label=f'Original {column} ({original_rate} Hz)')
    plt.plot(resampled_df.index, resampled_df[column], 'x-', label=f'Resampled {column} ({new_rate} Hz)')
plt.xlabel('Time (seconds)')
plt.ylabel('Value')
plt.legend()
plt.title(f'Resampling from {original_rate} Hz to {new_rate} Hz')
plt.show()
# Print resampled DataFrame
print(resampled_df)
```
